```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("autosdataset.xlsx")
```

```python
print("Estadísticas descriptivas básicas y medidas de tendencia central:")
print("Valor mínimo de precios:", df['price'].min())
print("Valor máximo de precios:", df['price'].max())
print("Moda de precios:", df['price'].mode()[0])
print("Media de precios:", df['price'].mean())
print("Mediana de precios:", df['price'].median())
print("Desviación estándar de precios:", df['price'].std())
print("Estadísticos y cuartiles:")
print(df['price'].describe())
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins='auto', color='grey', alpha=0.7)
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Histograma de Precios de Automóviles')
plt.grid(True)
plt.show()
```

```python
num_bins = int(np.ceil(1 + np.log2(len(df))))
print("Cantidad óptima de bins:", num_bins)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='body-style', y='price', data=df)
plt.xlabel('Tipo de Vehículo')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Tipo de Vehículo')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.xlabel('Tipo de Tracción')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Tipo de Tracción')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='engine-location', y='price', data=df)
plt.xlabel('Ubicación del Motor')
plt.ylabel('Precio')
plt.title('Relación entre Precio y Ubicación del Motor')
plt.grid(True)
plt.show()
```
