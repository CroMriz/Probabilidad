```python
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom
```

# Problema 1: Distribucion de Poisson

# Este problema se resuelve utilizando la distribucion de Poisson porque estamos tratando con eventos que ocurren independientemente en un intervalo de tiempo fijo (ingresos diaros a la UCI). La media es de 5 personas por dia

```python
media = 5

# a)

prob_2 = poisson.pmf(2, media)
print(f"La probabilidad de que el numero de personas ingresadas sea dos es: {prob_2:.4f}")
```

```python
# b)

prob_excede_10 = 1 - poisson.cdf(10, media)
print(f"La probabilidad de que el numero de personas ingresadas exceda de 10 es: {prob_excede_10:.4f}")
```

```python
x = range(0, 20)
y = poisson.pmf(x, media)

plt.bar(x, y, color='blue', alpha=0.7)
plt.title('Distribucion de Poisson (media=5)')
plt.xlabel('numero de personas ingresadas')
plt.ylabel('Probabilidad')
plt.show()
```

# Problema 2: Distribuccion Binomial

# Este problema se resuelve utilizando la distribucion binomial porque estamos tratando con una serie de ensayos independientes, cada uno con dos posibles resultados (defectuoso o no defectuoso). La probabilidad de exito (defectuoso) es 0.001 y el numero de ensayos es 1000

```python
n = 1000
p = 0.001

# a)

prob_0_defectuoso = binom.pmf(0, n, p)
print(f"La probabilidad de que ninguno sea defectuoso es: {prob_0_defectuoso:.4f}")
```

```python
# b)

prob_3_defectuosos = binom.pmf(3, n, p)
print(f"La probabilidad de que tres sean defectuosos es: {prob_3_defectuosos:.4f}")
```

```python
# c)

prob_4_defectuosos = binom.pmf(4, n, p)
print(f"La probabilidad de que cuatro sean defectuosos es: {prob_4_defectuosos:.4f}")
```

```python
x_binom = range(0, 15)
y_binom = binom.pmf(x_binom, n, p)

plt.bar(x_binom, y_binom, color='blue', alpha=0.7)
plt.title('Distribucion Binomial (n=1000, p=0.001)')
plt.xlabel('numero de motores defectuosos')
plt.ylabel('Probabilidad')
plt.show()
```
