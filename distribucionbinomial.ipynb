```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desv_std = 4
num_datos = 500

datos = np.random.normal(media, desv_std, num_datos)
datos[0:50]
```

```python
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))
```

```python
print("imprimir los primeros 50 valores de x", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucion normal ')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Pesos en Kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )

plt.title('Distribucion Normal ')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 79

prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)

estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

media = 70
desv_std = 4
peso_limite = 79
total_estudiantes = 500

prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

prob_mas_79kg = 1 - prob_menos_79kg

estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilidad de que un estudiante pese mas de 79 kg:", prob_mas_79kg)
print("Numero aproximado de estudiantes que pesan mas de 79 kg:", estudiantes_mas_79kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 79 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.figure(figsize=(6, 4))

media = 70
desv_std = 4
peso_limite1 = 68
peso_limite2 = 72
total_estudiantes = 500

prob_menos_68kg = norm.cdf(peso_limite1, media, desv_std)

prob_mas_68kg = 1 - prob_menos_68kg

estudiantes_mas_68kg = total_estudiantes * prob_mas_68kg

prob_menos_72kg = norm.cdf(peso_limite2, media, desv_std)

estudiantes_menos_72kg = total_estudiantes * prob_menos_72kg

prob_entre_68_y_72 = norm.cdf(peso_limite2, media, desv_std) - norm.cdf(peso_limite1, media, desv_std)

estudiantes_entre_68_y_72 = total_estudiantes * prob_entre_68_y_72

print("Probabilidad de que un estudiante pese menos de 72 kg:", prob_menos_72kg)
print("Numero de estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)

print("Probabilidad de que un estudiante pese mas de 68 kg:", prob_mas_68kg)
print("Numero aproximado de estudiantes que pesan mas de 68 kg:", estudiantes_mas_68kg)

print("Probabilidad de que un estudiante pese más de 68 kg y menos de 72 kg:", prob_entre_68_y_72)
print("Número aproximado de estudiantes que pesan más de 68 kg y menos de 72 kg:", estudiantes_entre_68_y_72)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite1), color='skyblue', alpha=0.5)
plt.fill_between(x, y, where=(x <= peso_limite2), color='red', alpha=0.5)
plt.fill_between(x, y, where=(x >= peso_limite1) & (x <= peso_limite2), color='white', alpha=0.3)
plt.title('Probabilidad de que un estudiante pese más de 68 kg y menos de 72 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
